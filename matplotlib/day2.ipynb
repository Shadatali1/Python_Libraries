# **Line Plot (Deep Understanding + Multiple Lines + Styling)**

```python
import matplotlib.pyplot as plt
```

### What is a Line Plot (Deep Concept)?

👉 A line plot shows how data changes over a continuous variable

### Key Idea

We use line plots to see trends over time or sequence

| Day | Sales |
| --- | ----- |
| 1   | 100   |
| 2   | 120   |
| 3   | 150   |

`👉 Question:`

Is sales increasing? ✔️\
How fast? ✔️

👉 This is why line plots are powerful.

### Basic Line Plot (Revision)

```python
days = [1,2,3,4]
sales = [100,120,150,170]

fig, ax = plt.subplots()
ax.plot(days, sales)

plt.show()
```

### Multiple Lines in One Graph (IMPORTANT)

👉 Real data science rarely has one variable\
👉 You compare things

```python
# Two Products Sales
days = [1,2,3,4]

product_A = [100,120,150,170]
product_B = [90,110,140,160]

fig, ax = plt.subplots()

ax.plot(days, product_A)
ax.plot(days, product_B)

plt.show()
```

**Problem here**

👉 Which line is A? Which is B? ❌\
👉 Confusing for users

### Legend (VERY IMPORTANT)

👉 Legend tells what each line represents

```python
fig, ax = plt.subplots()

ax.plot(days, product_A, label="Product A")
ax.plot(days, product_B, label="Product B")

ax.legend()

plt.show()
```

`Rule`

Always use legend when multiple lines exist

### Styling Basics (Make Graph Professional)
Add color + line style

```python
fig, ax = plt.subplots()

ax.plot(days, product_A, color='blue', linestyle='-', label="Product A")
ax.plot(days, product_B, color='red', linestyle='--', label="Product B")

ax.legend()

plt.show()
```

### Full Real-World Example

```python
days = [1,2,3,4,5]

product_A = [100,120,150,170,200]
product_B = [90,110,140,160,180]

fig, ax = plt.subplots()

ax.plot(days, product_A, label="Product A", color='blue')
ax.plot(days, product_B, label="Product B", color='green')

ax.set_title("Sales Comparison")
ax.set_xlabel("Days")
ax.set_ylabel("Sales")

ax.legend()

plt.show()
```

### Data Science Thinking (VERY IMPORTANT)

👉 What are we doing here?

We are:

- Comparing two variables
- Finding trends
- Making decisions

**Example Insight**\
👉 If Product A > Product B\
→ Focus marketing on A

👉 If B growing faster\
→ Future opportunity

### Mini Statistics Concept

👉 What are we comparing?

- Same X-axis (time)
- Two numerical variables

👉 This is called:

> Trend Comparison
