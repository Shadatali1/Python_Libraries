### Problem with Basic Plots

👉 Your current plots are correct\
❌ But not professional enough

**Example issue:**

- Hard to see exact data points
- Not visually attractive
- Not clear for presentation

### Markers (VERY IMPORTANT)

👉 Markers show each data point clearly

```python
import matplotlib.pyplot as plt

days = [1,2,3,4]
sales = [100,120,150,170]

fig, ax = plt.subplots()

ax.plot(days, sales, marker='o')

plt.show()
```

**What changed?**

👉 Each point is now visible\
👉 Easier to understand actual values

**Common Markers:**

| Marker | Meaning  |
| ------ | -------- |
| 'o'    | circle   |
| 's'    | square   |
| '^'    | triangle |
| '*'    | star     |


### Combine Marker + Line Style

```python
fig, ax = plt.subplots()

ax.plot(days, sales, marker='o', linestyle='--')

plt.show()
```

Now your graph is:

- Informative ✔️
- Visually clear ✔️

### Multiple Lines with Markers

```python
days = [1,2,3,4,5]

product_A = [100,120,150,170,200]
product_B = [90,110,140,160,180]

fig, ax = plt.subplots()

ax.plot(days, product_A, marker='o', label="Product A")
ax.plot(days, product_B, marker='s', label="Product B")

ax.legend()

plt.show()
```

### Why Markers Matter (Data Science Thinking)

👉 Without markers:

- You see trend only

👉 With markers:

- You see exact observations

**Example Insight**

👉 If day 3 suddenly drops:

- Marker shows exact point
- Helps detect anomalies

### Line Width (Professional Look)

```python
fig, ax = plt.subplots()
ax.plot(days, product_A, linewidth=8)
plt.show()
```

👉 Thicker line = better visibility

### Grid (VERY IMPORTANT for readability)

```python
fig, ax = plt.subplots()
ax.plot(days, product_A, linewidth=8)

ax.grid(True)
plt.show()
```

👉 Helps:

- Read values easily
- Looks professional

### Full Professional Example

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]

product_A = [100,120,150,170,200]
product_B = [90,110,140,160,180]

fig, ax = plt.subplots()

ax.plot(days, product_A, marker='o', linewidth=2, label="Product A")
ax.plot(days, product_B, marker='s', linewidth=2, label="Product B")

ax.set_title("Sales Comparison")
ax.set_xlabel("Days")
ax.set_ylabel("Sales")

ax.legend()
ax.grid(True)

plt.show()
```

### Visualization Rule (VERY IMPORTANT)

> A good graph should answer questions without explanation

Ask yourself:

```python
import matplotlib.pyplot as plt

# Data
days = [1,2,3,4,5,6,7]
study_hours = [2,3,4,5,6,5,7]
sleep_hours = [7,6,6,5,5,6,4]

# Create figure and axis
fig, ax = plt.subplots()

# Plot study hours
ax.plot(days, study_hours, marker='o', linestyle='-', label='Study Hours')

# Plot sleep hours
ax.plot(days, sleep_hours, marker='s', linestyle='--', label='Sleep Hours')

# Add title and labels
ax.set_title("Study vs Sleep Hours (Weekly Analysis)")
ax.set_xlabel("Days")
ax.set_ylabel("Hours")

# Add grid
ax.grid(True)

# Add legend
ax.legend()

# Show plot
plt.show()
```
