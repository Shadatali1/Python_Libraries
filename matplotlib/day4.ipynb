# **Scatter Plot (Relationships + Correlation)**

```python
import matplotlib.pyplot as plt
```

### What is a Scatter Plot?

👉 **Simple definition:**

A scatter plot shows the relationship between two numerical variables

**Important Condition**

👉 BOTH variables must be:

- Numerical ✔️
- Continuous (mostly) ✔️

❌ **Wrong Use**
- Category + number → use bar chart
- Only one variable → use histogram

### First Scatter Plot

```python
import matplotlib.pyplot as plt

height = [150, 160, 170, 180, 190]
weight = [50, 60, 70, 80, 90]

fig, ax = plt.subplots()

ax.scatter(height, weight)

ax.set_xlabel("Height")
ax.set_ylabel("Weight")
ax.set_title("Height vs Weight")

plt.show()
```

### What Are We Seeing?

👉 Each point = one person

Example:

(150, 50)\
(160, 60)

**Key Idea**

> Scatter plot shows how one variable changes with another

### Correlation (VERY IMPORTANT 🔥)

👉 **Correlation means:**

> How strongly two variables are related

**Types of Correlation**

✅ Positive Correlation

👉 As X increases → Y increases

Example:

Height ↑ → Weight ↑

---

❌ Negative Correlation

👉 As X increases → Y decreases

Example:

Study ↑ → Sleep ↓

---
⚖️ No Correlation

👉 No clear pattern

Example:

Shoe size vs exam marks

### Styling Scatter Plot

```python
fig, ax = plt.subplots()

ax.scatter(height, weight, color='blue')

ax.grid(True)

plt.show()
```

### Marker Size (s parameter)

```python
fig, ax = plt.subplots()

ax.scatter(height, weight, color='blue' , s = 100)

ax.grid(True)

plt.show()
```

### Color Based on Data (Advanced Intro)

```python
fig, ax = plt.subplots()

colors = [10,20,30,40,50]
ax.scatter(height, weight , s = 100 ,c = colors)

ax.grid(True)

plt.show()
```

**Used for:**

Third variable visualization

### Real-World Example
Study vs Marks

```python
study_hours = [1,2,3,4,5,6,7]
marks = [40,50,55,65,70,80,90]

fig, ax = plt.subplots()
colors = [10 , 20 , 30 , 40 , 50 , 60 , 70]
ax.scatter(study_hours, marks, c = colors , s = 150 , marker='^')

ax.set_title("Study Hours vs Marks")
ax.set_xlabel("Study Hours")
ax.set_ylabel("Marks")

ax.grid(True)

plt.show()
```

### Data Science Thinking

👉 Questions you should ask:

- Is there a relationship?
- Is it strong or weak?
- Any outliers?

### Example Insight

👉 If one point is far away:

- That’s an outlier
- Important in real analysis
