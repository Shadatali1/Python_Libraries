### Your First Plot (Basic Way)

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.show()
```

👉 This creates a simple line chart

### CORE CONCEPT — Figure vs Axes (VERY IMPORTANT)
`This is the foundation of everything in Matplotlib.`

**Real-Life Analogy**
Think like this:

- Figure = Entire page / canvas
- Axes = The area where graph is drawn

### Professional Code (Correct Way)

```python
fig, ax = plt.subplots()

ax.plot([1,2,3], [4,5,6])

plt.show()
```

| Element   | Meaning                            |
| --------- | ---------------------------------- |
| fig       | Full figure (container)            |
| ax        | Graph area (where data is plotted) |
| ax.plot() | Draws the graph                    |

### Real-World Example
Temperature Over Days

```python
days = [1,2,3,4,5]
temp = [30, 32, 35, 33, 31]

fig, ax = plt.subplots()

ax.plot(days, temp)

ax.set_title("Temperature Over Days")
ax.set_xlabel("Days")
ax.set_ylabel("Temperature")

plt.show()
```

**What did you learn here?**
- X-axis → Days
- Y-axis → Temperature
- You can see trend (increase/decrease)
